# World Happiness Analysis

### This dataset focuses on the happiness metrics of each country, and along with it some other world health statistics

## Introduction

World happiness is difficult to interpret and predict as every country is drastically different from each other and there is no concrete measure of happiness. This project seeks to quantify happiness in such a way that it can be analyzed to determine what makes some countries happier than others. 

The goal is to find statistics about various countries, and then use several techniques to assess the relationship between these statistics and the happiness of the countries.

We found a dataset created by the World Happiness Report that essentially assigns a ladder score to each country (signifying its happiness score). This is what we are trying to maximize. It also reveals several related attributes for the country (freedom, corruption, etc). Furthermore, we looked at other interesting world statistics to see if there was a relationship between those and world happiness as well. We found datasets on government expense on education, population, and government expense on research and development and tied those into the happiness dataset. Clearly, we take into account a lot of different statistics about countries around the world.

Given all this information about a country, the models we create will find the relationship between them and the ladder score. These models will analyze the data in different ways and suggest a corresponding ladder score, which we can then compare to the true ladder score. For example, we create linear regression models and random forest regressor models. The accuracy of each of these models reveals something about which factors are the most instrumental in determining the happiness of a country, and what countries can focus on to improve future happiness.

## Research Questions

* What factors contribute to a country's overall happiness?
* What differentiates happy countries from unhappy countries?
* Is there a correlation between happiness and other world statistics?

## Data Description

#### Descriptions of observations and attributes

The rows correspond to different countries in the happiness dataset, and the different factors that affect its quantifiable happiness levels. These factors (which are our columns) include the regional indicator, the ladder score (essentially our happiness score), the Gross Domestic Product Per Capita, social support indicators, life expectancy, percevied freedom to make life choices, perceived level of corruption, and the demographic's generosity. The details of each of these attributes, which their respective column names in the DataFrame, are listed below:
- Country or region (country): The country the data is being reported for
- Regional Indicator (r_indicator): The region is country belongs to (this is different from the continent)
- Score (l_score): The survey measure of SWB is from the Feb 28, 2020 release of the Gallup World Poll (GWP) covering years from 2005 to 2019.
- GDP per capita (gdp_per_capita): The statistics of GDP per capita (variable name gdp) in purchasing power parity (PPP) at constant 2011 international dollar prices are from the November 28, 2019 update of the World Development Indicators (WDI).
- Social support (support): The national average of the binary responses (either 0 or 1) to the GWP question “If you were in trouble, do you have relatives or friends you can count on to help you whenever you need them, or not?”
- Healthy life expectancy (life_expectancy): Healthy life expectancies at birth, which are based on the data extracted from the World Health Organization’s (WHO) Global Health Observatory data repository. 
- Freedom to make life choices (freedom): The national average of responses to the GWP question “Are you satisfied or dissatisfied with your freedom to choose what you do with your life?”
- Generosity (generosity): the residual of regressing national average of response to the GWP question “Have you donated money to a charity in the past month?” on GDP per capita.
- Perceptions of corruption (corruption): The measure is the national average of the survey re- sponses to two questions in the GWP: “Is corruption widespread throughout the government or not” and “Is corruption widespread within businesses or not?” The overall perception is just the average of the two 0-or-1 responses. In case the perception of government corruption is missing, we use the perception of business corruption as the overall perception. The corruption perception at the national level is just the average response of the overall perception at the individual level.

For more information, look at the [following file](files/Statistical_Appendix.pdf) (this is present in the project directory).

#### Processes might have influenced what data was observed and recorded and what was not
The rankings World Happiness Report 2020 use data that come from the Gallup World Poll (for more information see the Gallup World Poll methodology: https://www.gallup.com/178667/gallup-world-poll-work.aspx). The rankings are based on answers to the main life evaluation question asked in the poll, and the surveyees are sampled from a representative population that lead to a 95% confidence interval in the measurements reported. Gallup weights World Poll samples to correct for unequal selection probability.

The survey results might have been influenced by the surveyors’ ability to interact with the ‘ideal’ demographic. This might not be possible in countries struck by war, or intense government regulation. Furthermore, it is often the case that people in these countries do not report scores accurately. Moreover, different demographics often have different beliefs and ‘scales’ about the measurements being made. For instance, a rating of 4 on the Corruption scale for someone from Country X may equate to a rating of  7 for someone from Country Y. That being said, the organization is a reputed one, and they have made several efforts to accurately quantify metrics that are inherently very subjective. 

#### Concerns about people's privacy and their awareness of the motivation behind project
The randomly sampled people involved were aware of the data collection, as they were given a poll to answer from the Gallup World Poll, which is a well known organization. They expect the data to be used for analysis of various issues across the world, including food access, employment, leadership performance,  and well-being.

## Data Limitations

The data only gives the surface-level statistics on the select countries. There may be other underlying factors unaccounted for such as political conflicts, economic instability, and public health concerns that are not accounted for when analyzing the data.

Furthermore, the data used for this analysis focuses on 2019, which is not the most recent year. The 2020 data may be skewed due to the COVID-19 pandemic that took place during the majority of that year, so 2019 is the most recent year with accurate data; however, the same results may not still hold today.

Moreover, as previously explained, for our population, literacy rates, and government expenditure on education datasets, some countries lack data from recent years. The most recent data has been used for the analysis, but for some countries this may still be outdated informtion and not as accurate. This can be seen, for instance, in the data description for the government spending below - on average, the last data collection efforts for government education spending is from 2015. Of course, this is just a limitation of the dataset. As we rework and reformulate our hypothesis, we will manually search for this data for countries relevant to what we're doing.

## Linear Regression

Simple linear regression can be used to determine the relationship between the ladder score and other factors. A helper function for simplifying code is created to perform the linear regression and print out the slope and $r^2$ value

```python
my_linear_model = LinearRegression() 

def linearRegression(columns, y, train, test=None):
    for col in columns:
        my_linear_model.fit(train[col].values.reshape(-1, 1), train[y])
        slope = round(my_linear_model.coef_[0], 2)
        
        yhat_train = my_linear_model.predict(train[col].values.reshape(-1, 1))
        r2_train= round(r2_score(train[y], yhat_train), 2)
        train_values.append(r2_train)
        
        if test is not None:
            yhat_test = my_linear_model.predict(test[col].values.reshape(-1, 1))
            r2_test= round(r2_score(test[y], yhat_test), 2)
            test_values.append(r2_test)
        
        print("\n\n" + col)
        print('==================')
        print('Slope: ' + str(slope))
        print('r^2 value TRAIN: ' + str(r2_train))
        if test is not None:
            print('r^2 value TEST: ' + str(r2_test))
        
```

Linear regression is first performed with each of the six factors of the ladder score and the ladder score itself. For each one, the slope is printed out as well as the $r^2$ score, which indicates how much of a correlation there is.

For GDP per capita, support, and life expectancy, there is a positive slope with a somewhat high $r^2$ value, suggesting that a higher value of these generally means a happier country.

For generosity, there is a smaller positive slope with a near 0 $r^2$ value, suggesting that generosity is not too relevant in predicting ladder score.

For corruption, there is a larger positive slope but a relatively small $r^2$ value, suggesting that corruption increases a lot with ladder score, but there is not a huge correlation.

The train and test results are not too far apart from each other, suggesting that this is a decent linear model. The number of entries in the dataset are limited to the number of countries, which is a good amount of data to work with. If the dataset were larger though, test vs train accuracy could perhaps be improved.

```python
features = ['gdp_per_capita', 'support', 'life_expectancy', 'freedom', 'generosity', 'corruption']

train_values = []
test_values = []

linearRegression(features, 'l_score', train, test)
```

To determine the significance between the train and test values, we can use a t-test, which will compare the means of all the $r^2$ values. The p-value is about 0.5, which is at the midpoint of its range. This suggests that there may be a possibility these results occurred by chance (probably due to little data), but they are still generally accurate and not too different from each other.

```python
result = stats.ttest_ind(train_values, test_values)

print("statistic: " + str(result.statistic))
print("p-score: " + str(result.pvalue))

train_values = []
test_values = []
```

Furthermore, we can determine if there is a linear correlation between the ladder score of a country and the other statistics we are measuring here: population, literacy rates, and government expenditure on education. We first drop the rows that do not have a numeric value for the relevant columns, and use this modified dataset on this part.

The $r^2$ score for population is about 0, suggesting that there is no linear relationship between the population of a country and its happiness. For government expenditure on education, there is a small positive correlation (positive slope), suggesting that countries that spend more on education are slightly happier. Finally, for research there is an even higher positive correlation, suggesting that countries who spend more money on research and development are significantly happier.

```python
df2_train = train.dropna(subset = ['est_govt_education_spending', 'est_population', 'est_research'])
df2_test = test.dropna(subset = ['est_govt_education_spending', 'est_population', 'est_research'])

linearRegression(['est_govt_education_spending', 'est_population', 'est_research'],'l_score', df2_train, df2_test)
```

Once again we can perform a t-test to evaluate the signifance of these results. This has a much lower p-score, suggesting that these results are not by chance. The $r^2$ values are more different from each other compared to the previous model, so the the results are less accurate. There appears to be a tradeoff between the p-score and the accuracy in this dataset.

```python
result = stats.ttest_ind(train_values, test_values)

print("statistic: " + str(result.statistic))
print("p-score: " + str(result.pvalue))

train_values = []
test_values = []
```

This linear model looks at all of the countries equally. However, the results may differ depending on the region in the world we are looking at. As per the box plot above, the region with the highest happiness scores is North America, and the region with the lowest happiness scores is Sub-Saharan Africa. A linear model can be made specifically for these regions.

For North America, there are generally very high $r^2$ values (most of them near 1), suggesting that there is a very strong correlation between the six features of a ladder score and the ladder score itself.

The train/test split was not used in this case. This is because focusing on the countries from certain regions drastically reduces the sample size. Splitting the data even further would cause large variations due to lack of data.

```python
df_north_america = df[df['r_indicator'] == 'North America and ANZ']

linearRegression(features, 'l_score', df_north_america)
```

On the other hand, for Sub-Saharan Africa, there are very low $r^2$ values, suggesting that no matter the values of these features, these regions tended to have low ladder scores. 

This vast difference in $r^2$ values may mean that for countries that have a very low happiness score, there may be other factors that drop this value, as these six key features have little relevance. For countries that are overall very happy, variance in these six features is relevant in a slightly higher or lower happiness score.

## Multivariate Regression

Another helper function is created to perform multivariate regression with given features. A version is also created that does not print all the information, which will be used later on. 

```python
def multivariateRegression(features, y, df):
    my_linear_model.fit(df[features], df[y])

    print("coefficients:")
    print("-------------")
    for i in range(len(features)):
        print(features[i] + ": " + str(my_linear_model.coef_[i]))

    yhat = my_linear_model.predict(df[features])
    r2 = r2_score(df[y], yhat)
    print("\nr^2 value: " + str(r2))
```

```python
def multivariateRegressionNoPrinting(features, y, df):
    my_linear_model.fit(df[features], df[y])

    yhat = my_linear_model.predict(df[features])
    r2 = r2_score(df[y], yhat)
    return r2
```

Basic multivariate regression is performed using all of the six factors contributing to ladder score and ladder score itself. The slopes for each factor are printed out as well as the new $r^2$ score, which is a lot closer to one. This shows that a multivariate regression model is a better predictor of the happiness score than single linear regression.

```python
multivariateRegression(features, 'l_score', df)
```

One way we can approximate the true $r^2$ score between all of these factors and the ladder score is bootstrapping. With the bootstrap resampling method, we generate a 99% confidence interval for the $r^2$ score. 

The interval is from 0.71 to 0.87, and the approximated true $r^2$ score is 0.8, which is slightly closer to the higher end of the interval.

```python
n_bootstrap = 10000
bootstrap_values = []

for i in range(n_bootstrap):
    resampled_data = df.sample(replace=True, n=len(df))
    bootstrap_values.append(multivariateRegressionNoPrinting(features, 'l_score', resampled_data))
```

```python
quantiles = np.quantile(bootstrap_values, q=[0.005, 0.995])
low = quantiles[0]
high = quantiles[1]
print("confidence interval: [" + str(round(low, 2)) + ", " + str(round(high, 2)) + "]")

true_value = np.mean(bootstrap_values)
print("true r^2 value: " + str(round(true_value, 2)))
```

This model can still be improved upon by only selecting the features that are most relevant in predicting ladder score, so the statistics for each feature is described more in depth.

```python
X = np.column_stack((df['gdp_per_capita'], df['support'], df['life_expectancy'], df['freedom'], df['generosity'], df['corruption']))
y = df['l_score']
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

Looking at the p-values determine whether the variable is relevant to the model or not. x5 (generosity), has a large p-value, meaning that generosity is not statistically significant. The multivariate regression can be recomputed by removing this feature. Although the $r^2$ slightly drops, the predictions from this model would be a lot more accurate.

```python
features2 = ['gdp_per_capita', 'support', 'life_expectancy', 'freedom', 'corruption']

multivariateRegression(features2, 'l_score', df)
```

## Random Forest Regression

In the case that a linear model is not sufficient for analyzing ladder score, random forest regression can be used. This will create multiple decision trees using the data and then run points through the decision trees and take the average of all of its predictions. This starts with 5 estimators.

```python
rf = RandomForestRegressor(n_estimators = 5)
```

For decision trees to be created, all of the entries have to have a numerical value, so the columns that contain String values are removed. The labels are separated as well. The model is then fit and the predictions are computed.

```python
df3 = df.drop(['country', 'r_indicator', 'last_population_data', 'last_govt_spending_data', 'last_research_data'], axis = 1)
df3 = df3.dropna()
labels = np.array(df3['l_score'])
df3 = df3.drop('l_score', axis=1)
feature_list = list(df3.columns)
df3 = np.array(df3)
```

```python
rf.fit(df3, labels)
```

```python
predictions = rf.predict(df3)
```

The mean absolute error and mean absolute percentage error are computed to determine the accuracy. The accuracy is about 99%, which is very good. However, the number of estimators were set to 5 and the rest of the parameters were set to default, so this could perhaps do better.

```python
errors = abs(predictions - labels)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees')

mape = (errors / labels) * 100
accuracy = 100 - np.mean(mape)
print('Accuracy: ' + str(round(accuracy, 2)) + '%')
```

Hyperparameter tuning is performed using randomization. A grid is created of potential values, and random forest regression is performed several times, each with a different set of parameters. 

```python
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]
bootstrap = [True, False]

random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
rf_tuned = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
rf_tuned.fit(df3, labels)
```

The best parameters are noted, which are used to estimate the labels once again.

```python
rf_tuned.best_params_
```

```python
rf_tuned.best_estimator_
```

Once again, the errors and accuracy are computed. The accuracy improved by about 0.1%, so the hyperparameter tuning clearly provided a better random forest r.

```python
predictions_best = rf_tuned.predict(df3)
errors_best = abs(predictions_best - labels)
print('Mean Absolute Error:', round(np.mean(errors_best), 2), 'degrees')

mape_best = (errors_best / labels) * 100
accuracy_best = 100 - np.mean(mape_best)
print('Accuracy: ' + str(round(accuracy_best, 2)) + '%')

improve = round(accuracy_best - accuracy, 3)
print("\nImprovement: " + str(improve) + "%")
```

## Evaluation of Significance

Randomization was used, specifically in the case of random forest regression, for hyperparameter tuning. Given ranges, parameters were randomly chosen, and the accuracy was calculated to find the best set of parameters for the most accurate results.

For linear regression, a train/test split was used as well, where the model created with the training data was run with the test data to ensure similar results. Along with this, a t-test was performed to evaluate the significance of these differences.

For multivariate regression, bootstrap resampling was performed a large number of times to approximate a true $r^2$ value for the model.

## Conclusion

Both the simple linear regression model and the multivariate regression model suggest that generosity is the least correlated with ladder score. In the single regression model, generosity has the lowest $r^2$ score. In the multivariate model, predictions became a lot more accurate when generosity was not considered. 

Furthermore, for the other datasets, there appears to be a large correlation between research expenditure and ladder score. This suggests that countries who tend to invest more in scientific knowledge have happier citizens. On the other hand, population has a very low $r^2$ score, suggesting that the number of people in a country gives no indication of its happiness. 

For regions with generally high ladder scores (ie North America), small differences in the six factors have large variations in ladder scores. For example, having a slightly higher freedom score basically means having a higher overall ladder score. For regions with low ladder scores (ie Sub-Saharan Africa), variations in the six factors do not have much of an impact on ladder scores. For example, having a higher or lower corruption score than another country does not indicate if it is an overall happier country. This suggests that there may be factors outside of these six factors that could have a high correlation with ladder scores for these regions. Perhaps other datasets, similar to the government education spending, population, and government research spending, can be brought in to compare if these correlate to the ladder scores for these regions.

The random forest regression seemed to provide the most accurate results. This may be because randomization was used for tuning. There are a lot more factors here in constructing the model compared to the linear regression, such as the number of decision trees and the depth of decision trees, among many others. The accuracy, even without tuning, was approximately 99%, suggesting that this model is highly accurate in predicting ladder score based on the other factors.
